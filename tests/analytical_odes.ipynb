```python
from mdsim import ode_solver as ode

import sys

sys.path.append(".")
import analytical_systems as analsys
import statistics as stat

import numpy as np
import matplotlib.pyplot as plt
import matplotlib as mpl

mpl.rcParams["figure.dpi"] = 150
```

```python
sys1 = analsys.ode1()

x, t = ode.integrate(
    integrator=ode.Integrators.ODE_1_EXPLICIT_EULER,
    f=sys1.f,
    x0=[1],
    n_steps=10,
    tn=1,
)

x_anal = sys1.analytic_sol(t)
```

```python
# print(x, "\n", x_anal)
# print(t)
plt.plot(t, x)
plt.plot(t, x_anal)
```

```python
err = stat.relative_error(x.flatten(), x_anal)
err
```

```python
np.amax(err)
```

```python
np.logspace(np.amin(err), np.amax(err), 10)
```
